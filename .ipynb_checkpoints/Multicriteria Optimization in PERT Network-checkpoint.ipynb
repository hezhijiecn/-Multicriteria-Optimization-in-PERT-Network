Calculate the Adjacency Matrix

```python
def get_adjacency_matrix(project_data, sep,nan_value):
    ''' get the project activity ajacency matrix '''
    activity_name = project_data['Activity']            #get Activity name
    activity_number = project_data['Activity'].shape[0] #get Activity total number
    
    # initial the basic adjacency matrix
    adjacency_matrix = pd.DataFrame(np.zeros((activity_number,activity_number)),
                   index = list(activity_name),
                   columns=list(activity_name)
                   ).astype(int)
    # calculate the adjacency matrix 
    for index,activity in project_data.iterrows():
        for immediate_predecessor in activity[1].split(sep):
            if(immediate_predecessor != nan_value):
                adjacency_matrix[activity[0]][immediate_predecessor] = 1
                
    return adjacency_matrix
```

```python
adjacency_matrix = get_adjacency_matrix(project_data,sep, nan_value)
```

draw the project networks using networkx

```python
G = nx.DiGraph()

for index,activity in project_data.iterrows():
    
    G.add_node(activity[0],pos=(index,index+random.randint(0,14)))
    
    for immediate_predecessor in activity[1].split(sep):
        if(immediate_predecessor != nan_value):
            G.add_edge(immediate_predecessor, activity[0], weight=1)
            
#print('There are %d nodes in the graph with %d edges'%(G.number_of_edges(), G.number_of_nodes()))

pos=nx.get_node_attributes(G,'pos')
nx.draw(G, pos, node_color='gold', edge_color='grey',with_labels=True, font_weight='bold')
```

Calculate the Earliest start time(ES), Earliest finish time(EF), Latest start time(LS), Latest finish time(LF) and Slack

```python
def get_pert_time(project_data, adjacency_matrix, activity_name,activity_number):
    '''Calculate the pert time table with columns Earliest start time,Earliest finish time,Latest start time,Latest finish time and Slack '''
    
    pert_time_columns = ['ES','EF','LS','LF','Slack']
    pert_time = pd.DataFrame(np.zeros((activity_number,len(pert_time_columns))),
                       index = list(activity_name),
                       columns=list(pert_time_columns)
                       ).astype(int)
    
    # initial the necessary variables
    exe = [] # hold the executing activities
    time = 0 # hold the project current time
    flag = 0 # the number of finished project
    activity_visit = pd.DataFrame(np.zeros((1,activity_number)),columns=list(activity_name)).astype(int) # the visited list  1 means visited 
    project_data_copy = project_data.copy() # to not change the raw project data
    project_time = pd.DataFrame(project_data_copy['Estimated_Duration'].values,index=list(activity_name)) # get the project time
    indegree = adjacency_matrix.sum()  # get the initial indegree for each activity
    adjacency_matrix_copy = adjacency_matrix.copy() # not change the raw adjacency_matrix
    
    while flag != activity_number:
        for i in activity_name:
            if activity_visit[i].values ==0 and indegree[i]==0 and i not in exe:
                # enter time
                pert_time.loc[i,'ES'] = time
                # add this activity to execution queue
                exe.append(i)
        # calculate the execution queue minmum time   
        min_time = min(project_time[0][exe])

        time = time + min_time

        project_time[0][exe] = project_time[0][exe] - min_time

        for idx, val in project_time.iterrows():
            if val.values == 0 and activity_visit[idx].values == 0:

                #exit time
                exe.remove(idx)

                # add the flag
                flag = flag + 1

                # mark this activity has been visited
                activity_visit[idx] = 1

                pert_time.loc[idx,'EF'] = time
                # remove this activity outdegree
                adjacency_matrix_copy.loc[idx] = 0
                # recalculate the indegree
                indegree = adjacency_matrix_copy.sum()
    
    return pert_time
```

```python
pert_time = get_pert_time(project_data, adjacency_matrix, activity_name,activity_number)
pert_time
```
